### STEP 5. Divide the dataset into features (x) and the target variable (y) for training the model.

```python
x = df.drop('Target',axis=1)
y = df['Target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=3)
```

### STEP 6.Feature Engineering
- For Encoding,Scaling,Feature creation.


**LDA -** Linear DiscriminantAnalysis

- It is a supervised machine learning technique.
- Its main goal is to maximise class separability.
- LDA works by finding a linear combination of features that best separates the classes.

**Process** -
- Apply standardisation on the x dataframe.
- selecting the n components for creating the LDA.


- To find the number of components required for LDA, use the formula.
- n_components = min(no.of features,no.of classes-1).
- EX:-

    - number of features = 64
    - number of classes = 10
    - min(64,10-1) => min(64,9)
    - 9 is the minimum, so we will select 9 components.

```python
# STEP 1- Standardize the data
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

- We do not use fit on x_test because we don't want  do not use fit on x_test because we don't to recalculate the mean and sd

```python
x_train.std()
```

```python
x_test.std()
```

```python
pd.DataFrame(x_train)
```

```python
 # Apply LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
LDA = LinearDiscriminantAnalysis(n_components=9)
x_train = LDA.fit_transform(x_train,y_train)
x_test = LDA.transform(x_test)
```

```python
x_train.shape
```

```python
x_test.shape
```

### STEP 7.Model Selection & Training: Choose algorithms based on problem type.


```python
# Train the model.
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
# Predicting the values with test data.
model_pred = model.predict(x_test)
```

### STEP 8: Model Evaluation  
- Assessing the performance of the model based on various parameters.


```python
from sklearn.metrics import *
```

```python
accuracy_score(y_test,model_pred)
```

- After building the ML model using the LDA data, we evaluated it and achieved " 94 % accuracy ".
